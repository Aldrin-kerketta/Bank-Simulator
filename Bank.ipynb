```python
class Bank:
    bank_name = "SBI"
    IFSC = "SBIN0005457"
    bank_address = "Gajapati Nagar MainRoad, Khodasingh, Berhampur - 760010"
    _active = False
    __accNo = None
    
    def __init__(self):
        print("Welcome Page")
        print("------------\n")
        
        print("Welcome to SBI")
        print("This is demo Banking Program in python implemented by OOP\n")
    
    def _isValidPhoneNumber(self,phnum):
        if len(phnum) == 10:
            con = sql.Connection("Bank.db")
            query = 'SELECT * FROM "Account_Details" WHERE "PhoneNo" = "{}";'.format(phnum)
            ret = con.execute(query)
            if ret.fetchone() == None:
                con.close()
                return True
            else:
                print("Phone Number already exists")
                con.close()
                return 0
        else:
            print("Phone number should be of 10 digit")
            return False
    
    def _isValidPassword(self,pwd):
        upper = 0
        lower = 0
        num = 0
        sym = 0
        
        if len(pwd)>=6 and len(pwd)<=32:
            for l in pwd:
                if l.isalpha():
                    if l.isupper():
                        upper += 1
                    else:
                        lower += 1
                elif l.isnumeric():
                    num += 1
                else:
                    sym += 1
            
            if upper!=0:
                if lower!=0:
                    if num!=0:
                        if sym!=0:
                            return True
                        else:
                            print("Password should contain atleast 1 Symbol")
                            return False
                    else:
                        print("Password should contain atleast 1 Digit")
                        return False
                else:
                    print("Password should contain atleast 1 Lower case letter")
                    return False
            else:
                print("Password should contain atleast 1 Upper case letter")
                return False
        else:
            print("Password length should be between 6 to 32 charachter")
            return False
    
    def _isLeapYear(self,year):
        if year%4 == 0 and year%100 == 0 and year%400 == 0:
            return True
        else:
            return False
        
    def _isValidDate(self,dob):
        lastDate = {"1":31, "2":28, "L2": 29, "3": 31, "4": 30 , "5":31, "6": 30, "7":31, "8":31, "9":30, "10":31, "11":30, "12":31}
        flag_mo = False
        flag_dt = False
        
        ipDate = dob.split("-")
        date = int(ipDate[0])
        month = int(ipDate[1])
        year = int(ipDate[2])
        
        if month <= 0 or month>12:
            print("Month out of range")
            return False
        else:
            flag_mo = True
            if month == 2:
                if self._isLeapYear(year):
                    if date > lastDate["L2"]:
                        print("Date out of range")
                        return False
                    else:
                        flag_dt = True
                else:
                    if date > lastDate[str(month)]:
                        print("Date out of range",str(month))
                        return False
                    else:
                        flag_dt = True
            else:
                if date > lastDate[str(month)]:
                    print("Date out of range",str(month))
                    return False
                else:
                    flag_dt = True
        
        if flag_dt and flag_mo:
            diff = datetime.now() - datetime(year,month,date)
            if diff.days >=1825:
                return True
            else:
                print("User must need to be atleast 5 year old to bank account")
                return False
        
    def register_user(self,uType="H"):
        if self._active:
            print("First Logout then try")
            return
        else:
            print("Registration Page")
            print("-----------------\n")
            today = datetime.now()
            
            curr_date = today.strftime("%d")            
            curr_month = today.strftime("%m")        
            curr_year = today.strftime("%Y")
            
            str_today = curr_year+curr_month+curr_date
            
            con = sql.Connection("Bank.db")
            query = 'SELECT "AccNo" FROM "Account_Details" WHERE "AccNo" LIKE "{}%";'.format(str_today)
            ret = con.execute(query)
            row = ret.fetchone()
            rows = ret.fetchall()
            accNo = None
            
            if len(rows)!=0 or row!=None:
                if len(rows) < 1:
                    accNo = int(row[0][8:])+1
                    accNo = str_today+"002"
                else:
                    accNo = int(rows[len(rows)-1][0][8:])+1
                    if accNo<10:
                        accNo = str_today+"00"+str(accNo)
                    elif accNo>9 and accNo<100:
                        accNo = str_today+"0"+str(accNo)
                    elif accNo>99 and accNo<1000:
                        accNo = str_today+str(accNo)
                    else:
                        print("\nToo much user registered for today")
                        con.close()
                        return 
            else:
                accNo = str_today+"001"
            
            con.close()
            
            accType = ""
            Bal = 0.00
            while accType not in ["C","S","L","Q"]:
                accType = input("Account Type (C/S/L): ")
                accType=accType.upper()
                if accType == "Q":
                    print("\nRegistration cancelled !!!")
                    return 
                elif accType == "C":
                    Bal = 500.00
                elif accType == "S":
                    Bal = 2000.00
                elif accType == "L":
                    Bal = float(input("Enter Loan Amount: "))
            
            holderName = input("Enter Your Full Name: ")
            
            holderSex = ""
            while holderSex not in ["M","F","O","Q"]:
                holderSex = input("Enter Your Gender (M/F/O): ")
                holderSex=holderSex.upper()
                if holderSex == "Q":
                    print("\nRegistration cancelled !!!")
                    return
            
            dob_flag = True
            while dob_flag:
                dob = input("Enter your date of birth (dd-mm-yyyy): ")
                if self._isValidDate(dob):
                    dob_flag = False
                elif dob == "Q" or dob == "q":
                    dob_flag = False
                    print("\nRegistration cancelled !!!")
                    return 
            
            ph_flag = True
            while ph_flag:
                phNum = input("Enter your phone number (must be 10 digit): ")
                if self._isValidPhoneNumber(phNum):
                    ph_flag = False
                elif phNum == "Q" or phNum == "q":
                    ph_flag = False
                    print("\nRegistration cancelled !!!")
                    return 
            
            pwd_flag = True
            while pwd_flag:
                pwd = input("Enter your password (must match validation condition): ")
                if self._isValidPassword(pwd):
                    pwd_flag = False
                elif pwd == "Q" or pwd == "q":
                    pwd_flag = False
                    print("\nRegistration cancelled !!!")
                    return 
            
            now = datetime.now()
            time = now.strftime("%H")+":"+now.strftime("%M")
            date = curr_year+"-"+curr_month+"-"+curr_date
            
            con = sql.Connection("Bank.db")
            query = 'INSERT INTO "Account_Details" VALUES("{}", "{}", "{}", "{}", "{}", "{}", "{}", "{}", {});'.format(accNo,accType,holderName,holderSex,dob,phNum,pwd,uType,Bal)
            con.execute(query)
            con.commit()
            query = 'INSERT INTO "Transactions"("AccNo", "Date", "Time", "Operation", "PrevBal", "Amount", "AvailBal", "Details") VALUES("{}", "{}", "{}", "{}", {}, {}, {}, "{}");'.format(accNo,date,time,"Credit",0.00,Bal,Bal,"Account Created")
            con.execute(query)
            con.commit()
            print("\n\nSuccessfully Registered with Account Number {}".format(accNo))
            con.close()
    
    def login(self):
        if self._active:
            print("You are currently logged")
            return
        else:
            print("Login Page")
            print("----------\n")
            phNum = input("Enter your phone number:")
            pwd = input("Enter your password: ")
            
            con = sql.Connection("Bank.db")
            query = 'SELECT "AccNo", "FullName", "PhoneNo", "Password" FROM "Account_Details" WHERE "PhoneNo"="{}" AND "Password"="{}";'.format(phNum,pwd)
            ret = con.execute(query)
            row = ret.fetchone()
            if row!=None:
                self._active = True
                self.__accNo = row[0]
                print("\nSuccessfully Logged In")
                print("Hello {}".format(row[1]))
            else:
                print("Invalid Phone Number or Password !!!")
            con.close()
            
    def logout(self):
        if self._active:
            self._active = False
            self.__accNo = None
            print("\nSuccessfully Logged Out")
        else:
            print("Currently you are not Logged")
            return 
    
    def my_details(self):
        if self._active:
            Atype = None
            sex = None
            utype = None
            
            con = sql.Connection("Bank.db")
            query = 'SELECT * FROM "Account_Details" WHERE "AccNo" = "{}";'.format(self.__accNo)
            ret = con.execute(query)
            row = ret.fetchone()
            if row!=None:
                print("Details Page")
                print("------------\n")
                print("    Bank Details")
                print("    ------------")
                print("    Name              : ",self.bank_name)
                print("    IFSC              : ",self.IFSC)
                print("    Address           : {}\n".format(self.bank_address))
                print("    User Details")
                print("    ------------")
                print("    Account Number    : ",row[0])
                if row[1] == "C":
                    Atype = "Current"
                elif row[1] == "S":
                    Atype = "Savings"
                elif row[1] == "L":
                    Atype = "Loan"
                print("    Account Type      : ",Atype)
                print("    Holder Name       : ",row[2])
                if row[3] == "M":
                    sex = "Male"
                elif row[3] == "F":
                    sex = "Female"
                elif row[3] == "O":
                    sex = "Others"
                print("    Gender            : ",sex)
                print("    Date of Birth     : ",row[4])
                print("    Phone Number      : ",row[5])
                if row[7] == "H":
                    utype = "Account Holder"
                elif row[7] == "O":
                    utype = "Single Window Operator"
                elif row[7] == "A":
                    utype = "Admin"
                print("    User Type         : ",utype)
                print("    Available Balance : ",row[8])
        else:
            print("Please login to see your details")
            return
    
    def change_password(self):
        if self._active:
            print("Change Password")
            print("---------------\n")
            old_pwd = input("Enter Old Password  : ")
            new_pwd = input("Enter new Password  : ")
            cnf_pwd = input("Confirm new Password: ")
            
            if new_pwd == cnf_pwd:
                con = sql.Connection("Bank.db")
                query = 'SELECT * FROM "Account_Details" WHERE "AccNo" = "{}" AND "Password" = "{}";'.format(self.__accNo,old_pwd)
                ret = con.execute(query)
                row = ret.fetchone()
                if row!=None:
                    query = 'UPDATE "Account_Details" SET "Password" = "{}" WHERE "AccNo"="{}";'.format(new_pwd,self.__accNo)
                    con.execute(query)
                    con.commit()
                    con.close()
                    print("\nPassword updated successfully")
                else:
                    con.close()
                    print("\nOld password not matched !!!")
                    return                 
            else:
                print("\nMismatch of new password !!!")
                return 
        else:
            print("You first need to login to perform this action")
        
    def deposit(self):
        if self._active:
            utype = None
            
            con = sql.Connection("Bank.db")
            query = 'SELECT "FullName", "UserType" FROM "Account_Details" WHERE "AccNo" = "{}";'.format(self.__accNo)
            ret = con.execute(query)
            row = ret.fetchone()
            if row!=None:
                utype = row[1]
            con.close()
            if utype == "O" or utype == "A":
                print("Deposit Page")
                print("------------")
                print("Operator Name : {} \n".format(row[0]))
                dep_accNo = input("Enter Account of Depositer: ")
                dep_amt = float(input("Enter Deposit Amount: "))
                
                today = datetime.now()
                curr_date = today.strftime("%d")            
                curr_month = today.strftime("%m")        
                curr_year = today.strftime("%Y")
                
                now = datetime.now()
                time = now.strftime("%H")+":"+now.strftime("%M")
                date = curr_year+"-"+curr_month+"-"+curr_date
                
                con = sql.Connection("Bank.db")
                query = 'SELECT "FullName", "CurrBal" FROM "Account_Details" WHERE "AccNo"="{}"'.format(dep_accNo)
                ret = con.execute(query)
                row = ret.fetchone()
                if row!=None:
                    updateBal = row[1]+dep_amt
                    
                    print("\nHolder Name     : ",row[0])
                    print("Current Balance : ",row[1])
                    print("Deposit Amount  : ",dep_amt)
                    print("Updated Balance : ",updateBal)
                    
                    query = 'INSERT INTO "Transactions"("AccNo", "Date", "Time", "Operation", "PrevBal", "Amount", "AvailBal", "Details") VALUES("{}", "{}", "{}", "{}", {}, {}, {}, "{}");'.format(dep_accNo,date,time,"Credit",row[1],dep_amt,updateBal,"Deposited in Bank")
                    con.execute(query)
                    query = 'UPDATE "Account_Details" SET "CurrBal" = {} WHERE "AccNo"="{}";'.format(updateBal,dep_accNo)
                    con.execute(query)
                    con.commit()
                    con.close()
                    print("\nSuccessfully Deposited")
                    
                else:
                    con.close()
                    print("\nNo Account found on this account number !!!")
                    return 
        else:
            print("Ask a single window operator to deposite")
    
    def withdrawl(self):
        if self._active:
            print("Withdrawl Page")
            print("--------------\n")
            
            amt = float(input("Enter Withdrawl Amount: "))
            
            today = datetime.now()
            curr_date = today.strftime("%d")            
            curr_month = today.strftime("%m")        
            curr_year = today.strftime("%Y")
            
            now = datetime.now()
            time = now.strftime("%H")+":"+now.strftime("%M")
            date = curr_year+"-"+curr_month+"-"+curr_date
                
            con = sql.Connection("Bank.db")
            query = 'SELECT "CurrBal" FROM "Account_Details" WHERE "AccNo" = "{}";'.format(self.__accNo)
            ret = con.execute(query)
            row = ret.fetchone()
            if row!=None:
                updateBal = row[0] - amt
                if updateBal >= 0:
                    print("\nCurrent Balance   : ",row[0])
                    print("Withdrawl Amount  : ",amt)
                    print("Updated Balance   : ",updateBal)
                    
                    query = 'INSERT INTO "Transactions"("AccNo", "Date", "Time", "Operation", "PrevBal", "Amount", "AvailBal", "Details") VALUES("{}", "{}", "{}", "{}", {}, {}, {}, "{}");'.format(self.__accNo,date,time,"Debit",row[0],amt,updateBal,"Withdrawl from Bank")
                    con.execute(query)
                    query = 'UPDATE "Account_Details" SET "CurrBal" = {} WHERE "AccNo"="{}";'.format(updateBal,self.__accNo)
                    con.execute(query)
                    con.commit()
                    con.close()
                    print("\nSuccessfully Withdrawled")
                else:
                    con.close()
                    print("Not enough Balance !!!")
                    return 
            
        else:
            print("You need to Login to use your credits")
    
    def moneyTransfer(self):
        if self._active:
            print("Money Transfer Page")
            print("-------------------\n")
            
            trf_accNo = input("Enter Payee\'s Account Number: ")
            trf_amt = float(input("Enter amount to transfer     : "))
            
            payerBal = 0
            PayeeName = None
            payeeBal = 0
            
            today = datetime.now()
            curr_date = today.strftime("%d")            
            curr_month = today.strftime("%m")        
            curr_year = today.strftime("%Y")
            
            now = datetime.now()
            time = now.strftime("%H")+":"+now.strftime("%M")
            date = curr_year+"-"+curr_month+"-"+curr_date
            
            con = sql.Connection("Bank.db")
            query = 'SELECT "CurrBal" FROM "Account_Details" WHERE "AccNo" = "{}";'.format(self.__accNo)
            ret = con.execute(query)
            row = ret.fetchone()
            if row!=None:
                payerBal = row[0]
                
            query = 'SELECT "FullName", "CurrBal" FROM "Account_Details" WHERE "AccNo" = "{}";'.format(trf_accNo)
            ret = con.execute(query)
            row = ret.fetchone()
            if row!=None:
                payeeName = row[0]
                payeeBal = row[1]
                
                updatedPayerBal = payerBal - trf_amt
                updatedPayeeBal = payeeBal + trf_amt
                if updatedPayerBal >= 0:
                    print("\nPayee\'s Name      : ",payeeName)
                    print("\nAvailable Balance : ",payerBal)
                    print("Transfer Amount   : ",trf_amt)
                    print("Updated Balance   : ",updatedPayerBal)
                    
                    query = 'INSERT INTO "Transactions"("AccNo", "Date", "Time", "Operation", "PrevBal", "Amount", "AvailBal", "Details") VALUES("{}", "{}", "{}", "{}", {}, {}, {}, "{}");'.format(self.__accNo,date,time,"Debit",payerBal,trf_amt,updatedPayerBal,"Transferred to Account "+trf_accNo)
                    con.execute(query)
                    query = 'INSERT INTO "Transactions"("AccNo", "Date", "Time", "Operation", "PrevBal", "Amount", "AvailBal", "Details") VALUES("{}", "{}", "{}", "{}", {}, {}, {}, "{}");'.format(trf_accNo,date,time,"Credit",payeeBal,trf_amt,updatedPayeeBal,"Transferred from Account "+self.__accNo)
                    con.execute(query)
                    query = 'UPDATE "Account_Details" SET "CurrBal" = {} WHERE "AccNo"="{}";'.format(updatedPayerBal,self.__accNo)
                    con.execute(query)
                    query = 'UPDATE "Account_Details" SET "CurrBal" = {} WHERE "AccNo"="{}";'.format(updatedPayeeBal,trf_accNo)
                    con.execute(query)
                    con.commit()
                    con.close()
                    print("\nMoney Transferred to Account Number {} successfully".format(trf_accNo))
                else:
                    con.close()
                    print("\nNot Enough Balance !!!")
                    return 
            else:
                con.close()
                print("Payee\'s Account Not Found !!!")
                return 
        else:
            print("You need to Login to use your credits")
    
    def my_transactions(self):
        if self._active:
            print("Transaction Details Page")
            print("------------------------\n")
            
            print("options:")
            print("1. All\t\t2. By Year\t3. By Year & Month\tq/Q to cancel")
            opt = "E"
            while opt not in ["1", "2", "3", "q", "Q"]:
                opt = input("Enter your option: ")
                
                if opt == "q" or opt == "Q":
                    print("\nTransaction Display Cancelled !!!")
                    return 
                
                elif opt == "1":
                    con = sql.Connection("Bank.db")
                    query = 'SELECT * FROM "Transactions" WHERE "AccNo" = "{}";'.format(self.__accNo)
                    ret = con.execute(query)
                    row = ret.fetchone()
                    if row!=None:
                        print("")
                        print("TXN ID  |DATE        |TIME   |OPRT    |PREV_BAL  |AMOUNT  |AVAIL_BAL  |DESCRIPTION")
                        print("---------------------------------------------------------------------------------")
                        
                        stri = []
                        for i in range(len(row)-1):
                            if i != 1:
                                stri.append(str(row[i])+"          ")
                        stri.append(str(row[len(row)-1]))
                        
                        print(stri[0][:6],end="  |")
                        print(stri[1][:10],end="  |")
                        print(stri[2][:5],end="  |")
                        print(stri[3][:6],end="  |")
                        print(stri[4][:8],end="  |")
                        print(stri[5][:6],end="  |")
                        print(stri[6][:9],end="  |")
                        print(stri[7])
                        
                        for i in ret:
                            stri=[]
                            for j in range(len(i)-1):
                                if j != 1:
                                    stri.append(str(i[j])+"          ")
                            stri.append(str(i[len(i)-1]))
                            
                            print(stri[0][:6],end="  |")
                            print(stri[1][:10],end="  |")
                            print(stri[2][:5],end="  |")
                            print(stri[3][:6],end="  |")
                            print(stri[4][:8],end="  |")
                            print(stri[5][:6],end="  |")
                            print(stri[6][:9],end="  |")
                            print(stri[7])
                                                
                    else:
                        print("No Transactions found !!!")
                        
                elif opt=="2":
                    yr = input("Enter Year: ")
                    con = sql.Connection("Bank.db")
                    query = 'SELECT * FROM "Transactions" WHERE "AccNo" = "{}" AND "Date" LIKE "{}%";'.format(self.__accNo,yr)
                    ret = con.execute(query)
                    row = ret.fetchone()
                    if row!=None:
                        print("")
                        print("TXN ID  |DATE        |TIME   |OPRT    |PREV_BAL  |AMOUNT  |AVAIL_BAL  |DESCRIPTION")
                        print("---------------------------------------------------------------------------------")
                        
                        stri = []
                        for i in range(len(row)-1):
                            if i != 1:
                                stri.append(str(row[i])+"          ")
                        stri.append(str(row[len(row)-1]))
                        
                        print(stri[0][:6],end="  |")
                        print(stri[1][:10],end="  |")
                        print(stri[2][:5],end="  |")
                        print(stri[3][:6],end="  |")
                        print(stri[4][:8],end="  |")
                        print(stri[5][:6],end="  |")
                        print(stri[6][:9],end="  |")
                        print(stri[7])
                        
                        for i in ret:
                            stri=[]
                            for j in range(len(i)-1):
                                if j != 1:
                                    stri.append(str(i[j])+"          ")
                            stri.append(str(i[len(i)-1]))
                            
                            print(stri[0][:6],end="  |")
                            print(stri[1][:10],end="  |")
                            print(stri[2][:5],end="  |")
                            print(stri[3][:6],end="  |")
                            print(stri[4][:8],end="  |")
                            print(stri[5][:6],end="  |")
                            print(stri[6][:9],end="  |")
                            print(stri[7])
                    else:
                        print("No Transactions found !!!")
                        
                elif opt=="3":
                    yr = input("Enter Year: ")
                    mo = input("Enter Month: ")
                    con = sql.Connection("Bank.db")
                    query = 'SELECT * FROM "Transactions" WHERE "AccNo" = "{}" AND "Date" LIKE "{}%{}%";'.format(self.__accNo,yr,mo)
                    ret = con.execute(query)
                    row = ret.fetchone()
                    if row!=None:
                        print("")
                        print("TXN ID  |DATE        |TIME   |OPRT    |PREV_BAL  |AMOUNT  |AVAIL_BAL  |DESCRIPTION")
                        print("---------------------------------------------------------------------------------")
                        
                        stri = []
                        for i in range(len(row)-1):
                            if i != 1:
                                stri.append(str(row[i])+"          ")
                        stri.append(str(row[len(row)-1]))
                        
                        print(stri[0][:6],end="  |")
                        print(stri[1][:10],end="  |")
                        print(stri[2][:5],end="  |")
                        print(stri[3][:6],end="  |")
                        print(stri[4][:8],end="  |")
                        print(stri[5][:6],end="  |")
                        print(stri[6][:9],end="  |")
                        print(stri[7])
                        
                        for i in ret:
                            stri=[]
                            for j in range(len(i)-1):
                                if j != 1:
                                    stri.append(str(i[j])+"          ")
                            stri.append(str(i[len(i)-1]))
                            
                            print(stri[0][:6],end="  |")
                            print(stri[1][:10],end="  |")
                            print(stri[2][:5],end="  |")
                            print(stri[3][:6],end="  |")
                            print(stri[4][:8],end="  |")
                            print(stri[5][:6],end="  |")
                            print(stri[6][:9],end="  |")
                            print(stri[7])
                    else:
                        print("No Transactions found !!!")
            
        else:
            print("You need to login to see your transactions")
```
